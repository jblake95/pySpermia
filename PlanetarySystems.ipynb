# Planetary Systems
Simulating known systems with REBOUND.
## Kepler 47
Here we attempt to simulate the Kepler 47 circumbinary system.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import rebound

sim = rebound.Simulation()

# published parameters - see Orosz et al. 2012
M_A = 1.043 #M_solar
R_A = 0.964 #R_solar
M_B = 0.362 #M_solar
R_B = 0.351 #R_solar

a_AB = 0.0836 #AU
e_AB = 0.0234 
omega_AB = 212.3 #deg
inc_AB = 89.34 #deg

a_b = 0.2956 #AU
e_b = 0.035 
inc_b = 1 #89.59 #deg

a_c = 0.989 #AU
e_c = 0.411 
inc_c = 1 #89.826 #deg

# add primary, secondary and planets
sim.add(m=M_A)
sim.add(m=M_B, a=a_AB, e=e_AB, omega=omega_AB, inc=inc_AB)
sim.add(a=a_b, e=e_b, inc=inc_b)
sim.add(a=a_c, e=e_c, inc=inc_c)

sim.move_to_com()
fig = rebound.OrbitPlot(sim,slices=True,color=True,unitlabel="[AU]",lim=1.5,limz=0.36)
```
